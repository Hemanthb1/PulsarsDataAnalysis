```python

import numpy as np
import matplotlib.pyplot as plt
PSR=np.loadtxt('C:\\Users\\MayHem\\pen3.txt')

d=[]                    #empty array
for i in range(598):
    u=PSR[10*i:512+(10*i),5]        #selecting every 512 elements and advancing by 10 elements
    psr_fft = np.fft.fft(u)
    freq = np.fft.fftfreq(len(psr_fft))
    d.append(np.abs(psr_fft)[1:]**2)   #appending obtained psr_fft values to empty array                  #array containing 300 rows of data
f=np.transpose(d) 
plt.imshow(f[1:256])

plt.ylim(1,256)
plt.xlim(0,)



plt.yscale('log')
plt.xlabel('Start pulse number',fontsize=15)
plt.ylabel('frequency(Hz)',fontsize=15)
locs_x,labels_x = plt.xticks()
labels_x = [0,1000,2000,300,400,5000,6000]
plt.xticks(locs_x,labels_x)
locs_y,labels_y= plt.yticks()
labels_y= [float(item)*10**(-3) for item in locs_y]
#labels_y = [0.0001,0.001,0.01,0.1,0.5] 
plt.yticks(locs_y,labels_y)
plt.xlabel('starting pulse number',fontsize=15)
plt.ylabel('frequency(Hz)',fontsize=15)
plt.rc('xtick', labelsize=10)    # fontsize of the tick labels
plt.rc('ytick', labelsize=10)
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


PSR=np.loadtxt('C:\\Users\\MayHem\\pen3.txt')
c = PSR[1400:2000 ,5]
PSR_new = []

for i in range(60):
    PSR_new.append(np.mean(c[10*i:10*i+10]))
psr_fft = np.fft.fft(PSR_new)

freq = np.fft.fftfreq(len(psr_fft))
plt.plot(np.abs(freq)[1:],(np.abs(psr_fft)[1:])**2)
plt.xlabel('frequency(Hz)',fontsize=15)
plt.ylabel('Power',fontsize=15)
plt.xscale('log')
plt.rc('xtick', labelsize=15)    # fontsize of the tick labels
plt.rc('ytick', labelsize=15)
plt.rcParams["figure.figsize"] = (5,3)


plt.show()
```
