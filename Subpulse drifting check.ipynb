```python
import numpy as np
import matplotlib.pyplot as plt
PSR=np.loadtxt('C:\\Users\\MayHem\\e3.txt')                                         #initializing empty array
psr_new=[]                    #initializing emapty array for matrix
for i in range(162):
    m=[0]*1024
    for j in range(40):
        u=PSR[1024*(j+40*i):1024*((j+1)+40*i),3]
        m=m+u
    n=m/40
                                           #n is averaged value of cumulative 10 pulses
    psr_new.append(n)
```

```python

plt.imshow(psr_new)
#plt.ylim(0,162)
#plt.title('2D plot of single pulse 10 pulse binning',fontsize=16)
plt.xlabel('Phase Bins',fontsize=15)
plt.ylabel('Pulse number',fontsize=15)
#locs,labels = plt.yticks()
#labels = [float(item)*40 for item in locs] # to set x-labels as start pulse numbers
#plt.yticks(locs,labels)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.colorbar()
plt.axvline(x=511)

o=np.mean(psr_new,axis=0)
plt.figure(figsize=(20,8))
plt.plot(o)
plt.show()
```
