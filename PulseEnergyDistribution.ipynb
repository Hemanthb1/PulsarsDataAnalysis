```python
import matplotlib.pyplot as plt
import numpy as np



PSR1=np.loadtxt('C:\\Users\\MayHem\\pe-epoch2.txt')
x = PSR1[:,5]
y = PSR1[:,6]
plt.hist(x, density=False, bins=30,histtype=u'step')
plt.hist(y, density=False, bins=30,color='r',histtype=u'step')
 
# density=False would make counts
#plt.xlim(-0.2,0.2)
plt.ylabel('Number')
plt.xlabel('Energy')
plt.show()
```

```python
from mpl_toolkits.axes_grid.inset_locator import (inset_axes, InsetPosition, mark_inset)
fig, ax1 = plt.subplots()

# The data.
ax1.hist(x, density=False, bins=30,histtype=u'step')
# The Einstein fit.
ax1.hist(y, density=False, bins=30,color='r',histtype=u'step')
ax1.set_xlabel('Energy')
ax1.set_ylabel('Number')

ax2 = plt.axes([1,0,1,1])
0# Manually set the position and relative size of the inset axes within ax1
ip = InsetPosition(ax1, [0.7,0.7,0.3,0.3])
ax2.set_axes_locator(ip)
# Mark the region corresponding to the inset axes on ax1 and draw lines
# in grey linking the two axes.
mark_inset(ax1, ax2, loc1=2, loc2=4, fc="none", ec='0.5')
ax2.hist(x, density=False, bins=30,histtype=u'step')
# The Einstein fit.
ax2.hist(y, density=False, bins=30,color='r',histtype=u'step')
ax2.set_xlim(-0.06,0.00)
```
